```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Others\Downloads\SALES.CSV")

sold_units = df["total units"]

plt.plot(sold_units, color='Blue' , marker = 'o', linestyle = 'solid', label = "sold units", linewidth=1)

plt.xlabel("months")
plt.ylabel("sold units")

plt.title("sales Trend of all Products")
plt.legend(loc="lower right")
plt.xticks(range(len(df)), df['months'], rotation=45)
plt.show()
```
