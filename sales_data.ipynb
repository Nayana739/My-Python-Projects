```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Others\Downloads\SALES.CSV")

data = df[["months", "pen", "book", "marke", "chair", "table", "pen stand"]]
           
plt.stackplot(data["months"], data["pen"], data["book"], data["marke"],
          data["chair"], data["table"], data["pen stand"],
          labels=["pen", "book", "marke", "chair", "table", "pen stand"])


plt.xlabel("months")
plt.ylabel("sold units")

plt.title("sales Trend of all Products")
plt.legend(loc="upper left")

plt.show()
```
