```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Others\Downloads\SALES.CSV")

data=df[["months","chair","table"]]
bar_width=0.30

chair_bar=plt.bar(data.index,data['chair'],bar_width,color='Violet',  label='chair')

table_bar=plt.bar(data.index+bar_width,data['table'],bar_width, color='green', label='table')

plt.xlabel("months")
plt.ylabel("sold units")

plt.title("sales Trend of chair and table Products")
plt.legend(loc="upper left")
plt.xticks(data.index+bar_width/2, data['months'])
plt.show()
```
