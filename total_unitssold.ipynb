```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\Others\Downloads\SALES.CSV")

plt.plot(df["months"],df['pen'],color='blue',marker='o',label="pen")
plt.plot(df["months"],df['book'],color='red',marker='o',label="book")
plt.plot(df["months"],df['marke'],color='green',marker='o',label="marke")
plt.plot(df["months"], df['chair'], color='purple',marker='o',label="chair")
plt.plot(df["months"], df['table'], color='yellow',marker='o',label="table")
plt.plot(df["months"], df['pen stand'], color='orange',marker='o',label="pen stand")

plt.xlabel("months")
plt.ylabel("sold units")

plt.title("sales Trend By Products")
plt.legend(loc="upper left")

plt.show()
```
